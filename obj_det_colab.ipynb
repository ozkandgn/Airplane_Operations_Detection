````
import cv2
import numpy as np
def image_processing(img):
    blur_img = cv2.fastNlMeansDenoisingColored(img,None,6,3,5,12)

    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    kernel = kernel * 0.75
    return cv2.filter2D(blur_img, -1, kernel)
````

````
#for image prediction
# Import packages
import os
import cv2
import numpy as np
import tensorflow as tf
import sys

# This is needed since the notebook is stored in the object_detection folder.
sys.path.append("..")

# Import utilites
from utils import label_map_util
from utils import visualization_utils as vis_util

# Name of the directory containing the object detection module we're using
MODEL_NAME = 'inference_graph'
IMAGE_NAME = 'test_color.jpg'

# Grab path to current working directory
CWD_PATH = os.getcwd()

# Path to frozen detection graph .pb file, which contains the model that is used
# for object detection.
PATH_TO_CKPT = os.path.join(CWD_PATH,MODEL_NAME,'frozen_inference_graph.pb')

# Path to label map file
PATH_TO_LABELS = os.path.join(CWD_PATH,'training','labelmap.pbtxt')

# Path to image
PATH_TO_IMAGE = os.path.join(CWD_PATH,IMAGE_NAME)

# Number of classes the object detector can identify
NUM_CLASSES = 11

# Load the label map.
# Label maps map indices to category names, so that when our convolution
# network predicts `5`, we know that this corresponds to `king`.
# Here we use internal utility functions, but anything that returns a
# dictionary mapping integers to appropriate string labels would be fine
label_map = label_map_util.load_labelmap(PATH_TO_LABELS)
categories = label_map_util.convert_label_map_to_categories(label_map, max_num_classes=NUM_CLASSES, use_display_name=True)
category_index = label_map_util.create_category_index(categories)
print("ct",category_index)
# Load the Tensorflow model into memory.
detection_graph = tf.Graph()
with detection_graph.as_default():
    od_graph_def = tf.GraphDef()
    with tf.gfile.GFile(PATH_TO_CKPT, 'rb') as fid:
        serialized_graph = fid.read()
        od_graph_def.ParseFromString(serialized_graph)
        tf.import_graph_def(od_graph_def, name='')

    sess = tf.Session(graph=detection_graph)

# Define input and output tensors (i.e. data) for the object detection classifier

# Input tensor is the image
image_tensor = detection_graph.get_tensor_by_name('image_tensor:0')

# Output tensors are the detection boxes, scores, and classes
# Each box represents a part of the image where a particular object was detected
detection_boxes = detection_graph.get_tensor_by_name('detection_boxes:0')

# Each score represents level of confidence for each of the objects.
# The score is shown on the result image, together with the class label.
detection_scores = detection_graph.get_tensor_by_name('detection_scores:0')
detection_classes = detection_graph.get_tensor_by_name('detection_classes:0')

# Number of objects detected
num_detections = detection_graph.get_tensor_by_name('num_detections:0')

# Load image using OpenCV and
# expand image dimensions to have shape: [1, None, None, 3]
# i.e. a single-column array, where each item in the column has the pixel RGB value
image = cv2.imread(PATH_TO_IMAGE)
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY).reshape(360,640)
image[:,:,0]=gray_img
image[:,:,1]=gray_img
image[:,:,2]=gray_img
image_expanded = np.expand_dims(image, axis=0)

# Perform the actual detection by running the model with the image as input
(boxes, scores, classes, num) = sess.run(
    [detection_boxes, detection_scores, detection_classes, num_detections],
    feed_dict={image_tensor: image_expanded})
print("boxes=",boxes)
print("scores=",scores)
print("classes=",classes)
print("classes_type=",type(classes))
print("classes_0-0=",classes[0][0])
print("num=",num)

category_index[1]["name"] = "Aircraft Parked"
category_index[2]["name"] = "Chocks On"
category_index[3]["name"] = "Chocks On"#Second chocks value
category_index[4]["name"] = "Aft Starboard Cargo Door"
category_index[5]["name"] = "Aft Starboard Door"
category_index[6]["name"] = "Fwd Starboard Cargo Door"
category_index[7]["name"] = "Fwd Starboard Cargo Door"#Second FwdCargoDoor value
category_index[8]["name"] = "Fwd Starboard Door"
category_index[9]["name"] = "Bridge Connected"
category_index[10]["name"] = "Fueling"
category_index[11]["name"] = "Pushback Connected"

vis_util.visualize_boxes_and_labels_on_image_array(
    image,
    np.squeeze(boxes),
    np.squeeze(classes).astype(np.int32),
    np.squeeze(scores),
    category_index,
    use_normalized_coordinates=True,
    line_thickness=1,
    min_score_thresh=0.80,
    skip_labels = False,
    skip_scores = False)

from google.colab.patches import cv2_imshow

# All the results have been drawn on image. Now display the image.
cv2_imshow(image)
````

````
import time
import pandas as pd
import math
data = []  #for write csv 

def write_data(name,time_frame,is_start):
    if not is_start:
        if name == "Aircraft Parked":
            name = "Aircraft Parked Out"
        elif name == "Chocks On":
            name = "Chocks Off"
        elif name == "Bridge Connected":
            name = "Bridge Disconnected"
        elif name == "Aft Starboard Cargo Door":
            name = "Aft Starboard Cargo Door Closed"
        elif name == "Aft Starboard Door":
            name = "Aft Starboard Door Closed"
        elif name == "Fwd Starboard Cargo Door":
            name = "Fwd Starboard Cargo Door Closed"
        elif name == "Fwd Starboard Door":
            name = "Fwd Starboard Door Closed"
        elif name == "Fueling":
            name = "Fueling Ended"
        elif name == "Pushback Connected":
            name = "Pushback Started"
    total_time = int(time_frame / 30)
    minute = 0
    second = 0
    if total_time > 60:
        minute = math.floor(total_time / 60)
        second = total_time % 60
    else:
        second = total_time
    data.append([name,time.localtime().tm_year,time.localtime().tm_mon,time.localtime().tm_mday,time.localtime().tm_hour,time.localtime().tm_min + minute,time.localtime().tm_sec + second])
````

````
#for video
######## Video Object Detection Using Tensorflow-trained Classifier #########
#
# Import packages
import os
import cv2
import numpy as np
import tensorflow as tf
import sys

from google.colab.patches import cv2_imshow

# This is needed since the notebook is stored in the object_detection folder.
sys.path.append("..")

# Import utilites
from utils import label_map_util
from utils import visualization_utils as vis_util

# Name of the directory containing the object detection module we're using
MODEL_NAME = 'inference_graph'
VIDEO_NAME = 'test.mp4'

# Grab path to current working directory
CWD_PATH = os.getcwd()

# Path to frozen detection graph .pb file, which contains the model that is used
# for object detection.
PATH_TO_CKPT = os.path.join(CWD_PATH,MODEL_NAME,'frozen_inference_graph.pb')

# Path to label map file
PATH_TO_LABELS = os.path.join(CWD_PATH,'training','labelmap.pbtxt')

# Path to video
PATH_TO_VIDEO = os.path.join(CWD_PATH,VIDEO_NAME)

# Number of classes the object detector can identify
NUM_CLASSES = 11

# Load the label map.
# Label maps map indices to category names, so that when our convolution
# network predicts `5`, we know that this corresponds to `king`.
# Here we use internal utility functions, but anything that returns a
# dictionary mapping integers to appropriate string labels would be fine
label_map = label_map_util.load_labelmap(PATH_TO_LABELS)
categories = label_map_util.convert_label_map_to_categories(label_map, max_num_classes=NUM_CLASSES, use_display_name=True)
category_index = label_map_util.create_category_index(categories)

# Load the Tensorflow model into memory.
detection_graph = tf.Graph()
with detection_graph.as_default():
    od_graph_def = tf.GraphDef()
    with tf.gfile.GFile(PATH_TO_CKPT, 'rb') as fid:
        serialized_graph = fid.read()
        od_graph_def.ParseFromString(serialized_graph)
        tf.import_graph_def(od_graph_def, name='')

    sess = tf.Session(graph=detection_graph)

# Define input and output tensors (i.e. data) for the object detection classifier

# Input tensor is the image
image_tensor = detection_graph.get_tensor_by_name('image_tensor:0')

# Output tensors are the detection boxes, scores, and classes
# Each box represents a part of the image where a particular object was detected
detection_boxes = detection_graph.get_tensor_by_name('detection_boxes:0')

# Each score represents level of confidence for each of the objects.
# The score is shown on the result image, together with the class label.
detection_scores = detection_graph.get_tensor_by_name('detection_scores:0')
detection_classes = detection_graph.get_tensor_by_name('detection_classes:0')

# Number of objects detected
num_detections = detection_graph.get_tensor_by_name('num_detections:0')

# Open video file
video = cv2.VideoCapture(PATH_TO_VIDEO)


'''#didnt work
video.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
video.set(cv2.CAP_PROP_FRAME_HEIGHT, 360)
'''
out = cv2.VideoWriter('test_out.mp4',cv2.VideoWriter_fourcc('M','J','P','G'), 30, (640,360))
counter = 0

# %% personal codes for first last time%%

time_second = 2
treshold_second = 2
fps = 30

time_frame = time_second * fps
treshold_frame = treshold_second * fps

classes_first = np.full((12,1),True,dtype=bool)

timers = np.zeros((12,1),dtype=int)
timers_fps = np.zeros((12,1),dtype=int)
finding_times = np.zeros((12,1),dtype=int)
end_time = np.zeros((12,1),dtype=int)

# %% end personal codes for first last time %%

while(video.isOpened()):
    # Acquire frame and expand frame dimensions to have shape: [1, None, None, 3]
    # i.e. a single-column array, where each item in the column has the pixel RGB value
    ret, frame = video.read()
    frame = cv2.resize(frame,(640,360))
    '''
    gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY).reshape(360,640)
    frame[:,:,0]=gray_img
    frame[:,:,1]=gray_img
    frame[:,:,2]=gray_img
    '''
    frame_expanded = np.expand_dims(frame, axis=0)

    # Perform the actual detection by running the model with the image as input
    (boxes, scores, classes, num) = sess.run(
        [detection_boxes, detection_scores, detection_classes, num_detections],
        feed_dict={image_tensor: frame_expanded})
    
    fuel_values = np.where(classes[0] == 10.0)[0]
    if fuel_values.shape[0] != 0:
        clamp_index = fuel_values[0]
        if scores[0][clamp_index] > 0.35:
            scores[0][clamp_index] = 0.99
    aft_sd_values = np.where(classes[0] == 10.0)[0]
    if aft_sd_values.shape[0] != 0:
        clamp_index = aft_sd_values[0]
        if scores[0][clamp_index] > 0.70:
            scores[0][clamp_index] = 0.99
    # %% personal codes for first last time%%
  
    category_index[1]["name"] = "Aircraft Parked"
    category_index[2]["name"] = "Chocks On"
    category_index[3]["name"] = "Chocks On"#Second chocks value
    category_index[4]["name"] = "Aft Starboard Cargo Door"
    category_index[5]["name"] = "Aft Starboard Door"
    category_index[6]["name"] = "Fwd Starboard Cargo Door"
    category_index[7]["name"] = "Fwd Starboard Cargo Door"#Second FwdCargoDoor value
    category_index[8]["name"] = "Fwd Starboard Door"
    category_index[9]["name"] = "Bridge Connected"
    category_index[10]["name"] = "Fueling"
    category_index[11]["name"] = "Pushback Connected"

    
    len_scores = scores[scores>0.8].shape[0]
    
    classes_treshold = list(classes[0][:len_scores])
    classes_treshold = [int(i) for i in classes_treshold]
    
    timers[1:] -= 1
    timers[classes_treshold] += 3
    timers[timers<-treshold_frame/2] = -treshold_frame/2
    timers[timers>treshold_frame] = treshold_frame

    scores[:]=0.1
    for i in np.where(timers>=treshold_frame)[0]:
        if timers_fps[i] == 0:
            timers_fps[i] = treshold_frame
            finding_times[i] = counter
            #started
            write_data(category_index[i]["name"],counter,True)
            write_json(i,"open",int(counter/30))


    for i in np.where(timers_fps>1)[0]:
        #show
        timers_fps[i]-=1
        try:
            index = np.where(classes==float(i))[1][0]
        except:
            pass
        scores[0][index] = 0.99

    for i in np.where(timers_fps == 1)[0]:
        if timers[i] < 0:
            timers_fps[i] = -1
            end_time[i] = counter
            #ended
            write_data(category_index[i]["name"],counter,False)
            write_json(i,"close",int(counter/30))

    
    # %% end personal codes for first last time %%

    # Draw the results of the detection (aka 'visulaize the results')
    vis_util.visualize_boxes_and_labels_on_image_array(
        frame,
        np.squeeze(boxes),
        np.squeeze(classes).astype(np.int32),
        np.squeeze(scores),
        category_index,
        use_normalized_coordinates=True,
        line_thickness=1,
        min_score_thresh=0.8,
        skip_labels = False)
    
    out.write(frame)

    counter += 1
    if counter > 5000:
      break

# Clean up
video.release()
out.release()
cv2.destroyAllWindows()

print("finding:",finding_times)
print("end:",end_time)
print("data",data)
df = pd.DataFrame(data,columns=["Process","Year","Month","Day","Hour","Minute","Second" ])
df.to_csv("values.csv",encoding=("utf-8"),index=False)
with open('values.json', 'w') as outfile:
    json.dump(json_data, outfile)
````
